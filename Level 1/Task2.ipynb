```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Dataset .csv')
df.head()
df.describe()
# Top 10 cities
top_cities = df['City'].value_counts().head(10)
print(top_cities)

# Bar chart
top_cities.plot(kind='barh', figsize=(8,5), title='Top 10 Cities with Most Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.show()

# Top 10 cuisines
top_cuisines = df['Cuisines'].value_counts().head(10)
print(top_cuisines)

# Bar chart
top_cuisines.plot(kind='bar', figsize=(8,5), title='Top 10 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
# Frequency of each country code
df['Country Code'].value_counts().plot(kind='bar', figsize=(6,4), title='Country Code Distribution')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()


```
