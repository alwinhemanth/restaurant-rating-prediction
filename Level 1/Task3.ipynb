```python
import seaborn as sns
import matplotlib.pyplot as plt

# latitude vs longitude
plt.figure(figsize=(10,6))
sns.scatterplot(x='Longitude', y='Latitude', data=df, hue='Aggregate rating', palette='coolwarm', legend=False)
plt.title('Restaurant Locations by Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
# Number of restaurants in each city
city_counts = df['City'].value_counts().head(10)
print(city_counts)

# Bar chart
city_counts.plot(kind='barh', title='Top 10 Cities with Most Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.show()

# Average rating by city
avg_rating_by_city = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
print(avg_rating_by_city)

# Visualize
avg_rating_by_city.plot(kind='barh', title='Top 10 Cities by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('City')
plt.show()
# Correlation matrix
print(df[['Latitude', 'Longitude', 'Aggregate rating']].corr())
```
