```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Dataset .csv')
print(df['Has Table booking'].value_counts())
print(df['Has Online delivery'].value_counts())
print(df['Price range'].value_counts())
# Table Booking
table_booking_percent = df['Has Table booking'].value_counts(normalize=True) * 100
print("Table Booking %:\n", table_booking_percent)

# Online Delivery
online_delivery_percent = df['Has Online delivery'].value_counts(normalize=True) * 100
print("\nOnline Delivery %:\n", online_delivery_percent)

table_rating_comparison = df.groupby('Has Table booking')['Aggregate rating'].mean()
print("\nAverage Ratings (Table Booking):\n", table_rating_comparison)

# Optional Bar Plot
table_rating_comparison.plot(kind='bar', title='Average Rating by Table Booking Availability', color=['orange', 'green'])
plt.ylabel('Average Rating')
plt.show()

# Group by 'Price range' and 'Has Online delivery'
delivery_by_price = df.groupby(['Price range', 'Has Online delivery']).size().unstack()
print("\nOnline Delivery by Price Range:\n", delivery_by_price)

# Optional stacked bar chart
delivery_by_price.plot(kind='bar', stacked=True, title='Online Delivery Availability by Price Range')
plt.ylabel('Number of Restaurants')
plt.xlabel('Price Range')
plt.xticks(rotation=0)
plt.show()
```
