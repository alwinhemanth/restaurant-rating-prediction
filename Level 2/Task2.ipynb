```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/Dataset .csv')


print(df['Price range'].value_counts())
most_common_range = df['Price range'].mode()[0]
print("Most Common Price Range is:", most_common_range)
avg_rating_by_price = df.groupby('Price range')['Aggregate rating'].mean()
print("\nAverage Rating for Each Price Range:\n", avg_rating_by_price)

# Visualize it
avg_rating_by_price.plot(kind='bar', title='Avg Rating by Price Range', color='teal')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.show()
# Group by price range and rating color, then count
color_by_price = df.groupby(['Price range', 'Rating color']).size().unstack().fillna(0)
print("\nRating Color Frequency by Price Range:\n", color_by_price)

# Optional: stacked bar chart
color_by_price.plot(kind='bar', stacked=True, figsize=(8,5), title='Rating Color by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
# Group by price range and rating color, then count
color_by_price = df.groupby(['Price range', 'Rating color']).size().unstack().fillna(0)
print("\nRating Color Frequency by Price Range:\n", color_by_price)

# Optional: stacked bar chart
color_by_price.plot(kind='bar', stacked=True, figsize=(8,5), title='Rating Color by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.show()
```
