```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/Dataset .csv')


# Length of restaurant name
df['Name Length'] = df['Restaurant Name'].apply(len)
print(df[['Restaurant Name', 'Name Length']].head())
# Length of address
df['Address Length'] = df['Address'].apply(len)
print(df[['Address', 'Address Length']].head())
df['Has Table Booking (Encoded)'] = df['Has Table booking'].apply(lambda x: 1 if x.strip().lower() == 'yes' else 0)
print(df[['Has Table booking', 'Has Table Booking (Encoded)']].head())
df['Has Online Delivery (Encoded)'] = df['Has Online delivery'].apply(lambda x: 1 if x.strip().lower() == 'yes' else 0)
print(df[['Has Online delivery', 'Has Online Delivery (Encoded)']].head())
df.head()
```
